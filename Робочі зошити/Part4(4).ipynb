Варіант №20 **Кардіоїда**<br> 
Створити статичне зображення, на якому намалювати криву та супутні елементи. Зафіксувати стан у певний ненульовий момент часу.

```python
import numpy as np
import matplotlib.pyplot as plt
```

Задаємо параметри. Оскільки зображення динамічне, то зафіксувавши його в певний момент, задаються параметр кардіоїди (радіус твірного кола), а також момент часу (кут), для якого буде відображено статичний стан механізму.

```python
a = 5.0 
t_snap = 1.2 
```

Генеруємо дані для побудови кардіоїди за допомогою векторних операцій NumPy (без використання циклів). Використовуємо параметричні рівняння кардіоїди:

$$x = 2a(1 - \cos\varphi)\cos\varphi,\quad y = 2a(1 - \cos\varphi)\sin\varphi,\quad 0 \leq \varphi < 2\pi$$

```python
t_full = np.linspace(0, 2 * np.pi, 500)
x_cardioid = 2 * a * (1 - np.cos(t_full)) * np.cos(t_full)
y_cardioid = 2 * a * (1 - np.cos(t_full)) * np.sin(t_full)
```

Будуємо нерухоме твірне коло радіусом `a` з центром у точці (a, 0). Це коло демонструє геометричний спосіб утворення кардіоїди.

```python
theta_circle = np.linspace(0, 2 * np.pi, 200)
circle_center_x = a
circle_center_y = 0
circle_x = circle_center_x + a * np.cos(theta_circle)
circle_y = circle_center_y + a * np.sin(theta_circle)
```

Обчислюємо координати супутніх елементів для фіксованого моменту часу `t_snap`: положення генеруючої точки та радіус-вектор від центра до цієї точки.

```python
p_x = 2 * a * (1 - np.cos(t_snap)) * np.cos(t_snap)
p_y = 2 * a * (1 - np.cos(t_snap)) * np.sin(t_snap)

radius_x = [0, p_x]
radius_y = [0, p_y]
```

Візуалізуємо результат: будуємо кардіоїду, нерухоме коло, радіус-вектор, генеруючу точку, центр та додаємо анотації, легенду, сітку. Зберігаємо зображення у файл `plot_task4.png`.

```python
plt.figure(figsize=(10, 10))

plt.plot(circle_x, circle_y, color='black', linewidth=1.5,
         label=f'Твірне коло (радіус = a = {a})')

plt.plot(x_cardioid, y_cardioid, color='blue', linewidth=2.5,
         label='Кардіоїда')

plt.plot(radius_x, radius_y, color='gray', linestyle='--', linewidth=1.2,
         label='Радіус-вектор до точки P')

plt.scatter(p_x, p_y, color='red', s=100, zorder=10,
            label=f'Точка P (t = {t_snap:.2f})')

plt.scatter(0, 0, color='black', s=50, label='Центр O(0,0)')

plt.annotate(f'a = {a}', xy=(a/2, a/10),
             xytext=(a/2 + 1, a/10 + 1),
             arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=5))

plt.title(f'Кардіоїда (Варіант 20)\n$r = {a}(1 - \\cos\\varphi)$, момент $t_{{snap}} = {t_snap}$',
          fontsize=14, fontweight='bold')
plt.xlabel('X', fontsize=12)
plt.ylabel('Y', fontsize=12)

plt.legend(loc='upper right', fontsize=10)
plt.axis('equal')
plt.grid(True, linestyle=':', alpha=0.5)

plt.savefig('plot_task4.png', dpi=150, bbox_inches='tight')
plt.show()
```
