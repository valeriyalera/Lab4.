У завданні 4 лабораторної роботи №2 оброблялися значення функції у точках. Побудувати гістограму, яка відображає розподіл точок по інтервалах.

```python
import pandas as pd
import matplotlib.pyplot as plt
```


Завантажуємо експериментальні дані із CSV-файлу. Тут ми беремо всі точки до групування, щоб побачити реальний розподіл кількості проведених експериментів для різних розмірів вхідних даних (n).

```python
raw = pd.read_csv('lab2_var119.csv')

n_points = raw['n']
```

Будуємо гістограми розподілу значень.

```python
plt.figure(figsize=(10, 6))

plt.hist(n_points, bins=15, color='skyblue', edgecolor='black', alpha=0.8)

plt.title('Гістограма розподілу точок (розмірів даних $n$) по інтервалах',
          fontsize=14, fontweight='bold')
plt.xlabel('Розмір оброблюваних даних (n)', fontsize=12)
plt.ylabel('Кількість точок (експериментів) в інтервалі', fontsize=12)

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.savefig('plot_task3.png', dpi=150, bbox_inches='tight')
plt.show()
```
