Варіант № 8 **Пляшка Кляйна**
<br>Створити статичне зображення, на якому намалювати тривимірну поверхню (як wireframe plot). На зображенні має бути підпис (назва поверхні) та легенда (містить рівняння поверхні).


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

Задаємо параметризацію Пляшки Кляйна. Використовуємо стандартну параметричну форму:

$$x = \left(r + \cos\frac{u}{2}\sin v - \sin\frac{u}{2}\sin 2v\right)\cos u$$
$$y = \left(r + \cos\frac{u}{2}\sin v - \sin\frac{u}{2}\sin 2v\right)\sin u$$
$$z = \sin\frac{u}{2}\sin v + \cos\frac{u}{2}\sin 2v$$

де $u \in [0, 2\pi]$, $v \in [0, 2\pi]$, $r = 2$ — радіус "пляшки".

```python
r_klein = 2.0 

u = np.linspace(0, 2 * np.pi, 60)
v = np.linspace(0, 2 * np.pi, 60)
U, V = np.meshgrid(u, v)

x = (r_klein + np.cos(U/2) * np.sin(V) - np.sin(U/2) * np.sin(2*V)) * np.cos(U)
y = (r_klein + np.cos(U/2) * np.sin(V) - np.sin(U/2) * np.sin(2*V)) * np.sin(U)
z = np.sin(U/2) * np.sin(V) + np.cos(U/2) * np.sin(2*V)
```

Будуємо каркасну (wireframe) поверхню Пляшки Кляйна.

```python
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

ax.plot_wireframe(x, y, z, color='purple', linewidth=0.5, alpha=0.7)

equation_text = (r'$x = \left(r + \cos\frac{u}{2}\sin v - \sin\frac{u}{2}\sin 2v\right)\cos u$' + '\n' +
                 r'$y = \left(r + \cos\frac{u}{2}\sin v - \sin\frac{u}{2}\sin 2v\right)\sin u$' + '\n' +
                 r'$z = \sin\frac{u}{2}\sin v + \cos\frac{u}{2}\sin 2v$' + '\n' +
                 r'$r=2,\ u\in[0,2\pi],\ v\in[0,2\pi]$')

ax.plot([], [], [], ' ', label=equation_text)
ax.legend(loc='upper left', handlelength=0, fontsize=8, frameon=True, title="Рівняння поверхні")

ax.set_title("Пляшка Кляйна\n(Неорієнтовна поверхня без краю)", fontsize=14, fontweight='bold')
ax.set_xlabel('X', fontsize=11)
ax.set_ylabel('Y', fontsize=11)
ax.set_zlabel('Z', fontsize=11)

plt.savefig('plot_task5.png', dpi=150, bbox_inches='tight')
plt.show()
```
